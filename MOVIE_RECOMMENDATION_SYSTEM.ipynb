```python
# Selecting only the relevant features for recommendation
selected_features = ["genres", "keywords", "tagline", "cast", "director"]
```

```python
# Replacing the null values in selected features column with null string
for feature in selected_features:
  df[feature] = df[feature].fillna("")
```

```python
# Combining all the selected features
combined_features = df["genres"] + " " + df["keywords"] + " " + df["tagline"] + " " + df["cast"] + " " + df["director"]
```

```python
print(combined_features)
```

```python
# Converting the text data into numerical values
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

**USING COSINE SIMILARITY**

```python
# Getting the similarity score using cosine similarity
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
# Taking the input name of movie from the user
movie_name = input("Enter the name of your favourite movie: ")

print("My favourite movie is :", movie_name)
```

```python
# Creating a list with all movie names given in the dataset
all_movie_names = df["title"].tolist()
print(all_movie_names)
```

```python
# Finding the close match of the movie name taken as input from the user
find_close_match = difflib.get_close_matches(movie_name, all_movie_names)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# Finding the index of the movie with title
movie_index = df[df.title == close_match]["index"].values[0]
print(movie_index)
```

```python
# Getting a list of similar movies based on a given movie
similarity_score = list(enumerate(similarity[movie_index]))
print(similarity_score)
```

```python
# Sorting similar movies in terms of similarity score in descending order
sorted_similarity_score = sorted(similarity_score, key=lambda x: x[1], reverse=True)
print(sorted_similarity_score)
```

```python
# Printing the names of similar movies based on similarity score
print("Movies suggested : \n")

i=1

for movie in sorted_similarity_score:
  index = movie[0]
  title_from_index = df[df.index == index]["title"].values[0]
  if(i<30):
    print(i, ".", title_from_index)
    i+=1
```

**Movie Recommendation System**

```python
# Taking the input name of movie from the user
movie_name = input("Enter the name of your favourite movie: ")

# Creating a list with all movie names given in the dataset
all_movie_names = df["title"].tolist()

# Finding the close match of the movie name taken as input from the user
find_close_match = difflib.get_close_matches(movie_name, all_movie_names)

# Getting the closest matching movie
close_match = find_close_match[0]

# Finding the index of the movie with title
movie_index = df[df.title == close_match]["index"].values[0]

# Getting a list of similar movies based on a given movie
similarity_score = list(enumerate(similarity[movie_index]))

# Sorting similar movies in terms of similarity score in descending order
sorted_similarity_score = sorted(similarity_score, key=lambda x: x[1], reverse=True)

# Printing the names of similar movies based on similarity score
print("Movies suggested : \n")

i=1

for movie in sorted_similarity_score:
  index = movie[0]
  title_from_index = df[df.index == index]["title"].values[0]
  if(i<30):
    print(i, ".", title_from_index)
    i+=1
```
