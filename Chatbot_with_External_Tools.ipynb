```python
import requests
from langchain.tools import BaseTool

class CustomGoogleSearchTool(BaseTool):
    name = "Google Search Tool"
    description = "Custom tool for Google Search using SerpAPI."

    def _run(self, query: str):
        api_key = "eccd40c2748193a9fdb010fae66b5d58275369bf160bb9cab0343061c1ddd406"
        params = {
            "q": query,
            "api_key": api_key,
            "engine": "google",
        }
        response = requests.get("https://serpapi.com/search", params=params)
        data = response.json()

        search_results = data.get("organic_results", [])
        result_texts = []

        for result in search_results:
            title = result.get("title")
            link = result.get("link")
            snippet = result.get("snippet")
            result_texts.append(f"Title: {title}\nLink: {link}\nSnippet: {snippet}\n")

        formatted_results = "\n".join(result_texts)
        return formatted_results

google_search_tool = CustomGoogleSearchTool()
output = google_search_tool.run("What is the current hot topic in the football transfer window?")
print(output)
```

```python
from langgraph.graph.message import add_messages
class State(TypedDict):
  messages:Annotated[list,add_messages]
```

```python
from langgraph.graph import StateGraph,START,END
```

```python
graph_builder = StateGraph(State)
```

```python
from langchain_groq import ChatGroq
from google.colab import userdata
groq_api_key = userdata.get("groq_api_key")
llm = ChatGroq(groq_api_key=groq_api_key, model="Gemma2-9b-It")
llm
```

# **FLOW OF THE MODEL**

image.png

```python
llm_with_tools = llm.bind(tools=tools)
```

```python
def chatbot(state:State):
  return {"messages" : [llm_with_tools.invoke(state["messages"])]}
```

```python
from langgraph.prebuilt import ToolNode, tools_condition
```

```python
graph_builder.add_node("chatbot", chatbot)
graph_builder.add_edge(START, "chatbot")
tool_node = ToolNode(tools=tools)
graph_builder.add_node("tools", tool_node)

graph_builder.add_conditional_edges(
    "chatbot",
    tools_condition
)

graph_builder.add_edge("tools", "chatbot")
graph_builder.add_edge("chatbot", END)
```

```python
graph = graph_builder.compile()
```

```python
from IPython.display import Image, display

try:
  display(Image(graph.get_graph().draw_mermaid_png()))
except Exception:
  pass
```

```python
user_input = "Please type your question."

events = graph.stream(
    {"messages" : [("user", user_input)]}, stream_mode="values"
)

for event in events:
  print(event["messages"])
```

```python
user_input = "Please type your question:"

# Simulate getting events
events = graph.stream(
    {"messages": [("user", user_input)]}, stream_mode="values"
)

# Loop through each event
for event in events:
    try:
        # Access the messages key
        if "messages" in event:
            # Extract the last message from the messages list
            last_message = event["messages"][-1]

            # Print the type of the last message for debugging
            print(f"Message type: {type(last_message)}")

            # Attempt to convert the last message to a string
            if hasattr(last_message, "__dict__"):
                # Convert the object attributes to a dictionary and then to a string
                message_str = str(last_message.__dict__)
            else:
                # Fallback if __dict__ is not available
                message_str = str(last_message)

            print(f"Message content: {message_str}")
    except Exception as e:
        print(f"An error occurred: {e}")
```
