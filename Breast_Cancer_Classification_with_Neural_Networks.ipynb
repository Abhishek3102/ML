```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

**Separating the features and target**

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Standardize the data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

**Building the Neural Network**

download.png

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# Setting up the layers for Neural Network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    # Here 2 is chosen bcoz only two classes, i.e., 0 and 1 are there. That number will change according to the number of classes we have.
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# Compiling the Neural Network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Training the Neural Network
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

# Visualizing the accuracy and loss


```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.ylabel("Accuracy")
plt.xlabel("Epochs")

plt.legend(['Training Data', 'Validation Data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Model Loss')
plt.ylabel("Loss")
plt.xlabel("Epochs")

plt.legend(['Training Data', 'Validation Data'], loc='upper right')
```

**Accuracy of the model on test data**

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

model.predict() gives the prediction probability of each class for that data point

```python
# Converting the prediction probability to class labels

# argmax provides the index of the highest value of the list
Y_predict_labels = [np.argmax(i) for i in Y_predict]
print(Y_predict_labels)
```

**Building the predictive system**

```python
input_data = (13.54,14.36,87.46,566.3,0.09779,0.08129,0.06664,0.04781,0.1885,0.05766,0.2699,0.7886,2.058,23.56,0.008462,0.0146,0.02387,0.01315,0.0198,0.0023,15.11,19.26,99.7,711.2,0.144,0.1773,0.239,0.1288,0.2977,0.07259)

# change the input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# Standardizing the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0] == 0):
  print("The tumor is Malignant")

else:
  print("The tumor is Benign")
```
