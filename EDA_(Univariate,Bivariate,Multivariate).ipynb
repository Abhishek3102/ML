```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv")
```

```python
df.head()
```

## Univariate Analysis

```python
df_setosa = df.loc[df["species"] == "setosa"]
```

```python
df_versicolor = df.loc[df["species"] == "versicolor"]
df_virginica = df.loc[df["species"] == "virginica"]
```

```python
plt.plot(df_setosa['sepal_length'], np.zeros_like(df_setosa['sepal_length']), 'o', label='Setosa', color='blue')
plt.plot(df_versicolor['sepal_length'], np.zeros_like(df_versicolor['sepal_length']), 'o', label='Versicolor', color='green')
plt.plot(df_virginica['sepal_length'], np.zeros_like(df_virginica['sepal_length']), 'o', label='Virginica', color='red')
plt.legend()
plt.show()
```

## Bivariate Analysis

```python
sns.FacetGrid(df, hue='species', height=5).map(plt.scatter, "sepal_length", "sepal_width").add_legend()
plt.show()
```

```python
sns.FacetGrid(df, hue='species', height=5).map(plt.scatter, "petal_length", "sepal_width").add_legend()
plt.show()
```

```python
sns.FacetGrid(df, hue='species', height=5).map(plt.scatter, "sepal_length", "petal_length").add_legend()
plt.show()
```

```python
sns.FacetGrid(df, hue='species', height=5).map(plt.scatter, "petal_width", "sepal_width").add_legend()
plt.show()
```

## Multivariate Analysis

```python
sns.pairplot(df, hue='species', height=4)
plt.show()
```
