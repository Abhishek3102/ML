```python
# One-hot encoding
df_one_hot_encoded = pd.get_dummies(df, columns=['Gender'], prefix='Gender')

print(df_one_hot_encoded)
```

```python
df_one_hot_encoded.head()
```

```python
gender_converted_to_numerical_form = {
     True: 1,
     False: 0
 }

df_one_hot_encoded['Gender_Female'] = df_one_hot_encoded['Gender_Female'].map(gender_converted_to_numerical_form)
df_one_hot_encoded['Gender_Male'] = df_one_hot_encoded['Gender_Male'].map(gender_converted_to_numerical_form)
```

```python
from colorspacious import cspace_converter

import matplotlib.pyplot as plt
import numpy as np

import matplotlib as mpl

cmaps = {}

gradient = np.linspace(0, 1, 256)
gradient = np.vstack((gradient, gradient))


def plot_color_gradients(category, cmap_list):
    # Create figure and adjust figure height to number of colormaps
    nrows = len(cmap_list)
    figh = 0.35 + 0.15 + (nrows + (nrows - 1) * 0.1) * 0.22
    fig, axs = plt.subplots(nrows=nrows + 1, figsize=(6.4, figh))
    fig.subplots_adjust(top=1 - 0.35 / figh, bottom=0.15 / figh,
                        left=0.2, right=0.99)
    axs[0].set_title(f'{category} colormaps', fontsize=14)

    for ax, name in zip(axs, cmap_list):
        ax.imshow(gradient, aspect='auto', cmap=mpl.colormaps[name])
        ax.text(-0.01, 0.5, name, va='center', ha='right', fontsize=10,
                transform=ax.transAxes)

    # Turn off *all* ticks & spines, not just the ones with colormaps.
    for ax in axs:
        ax.set_axis_off()

    # Save colormap list for later.
    cmaps[category] = cmap_list


plot_color_gradients('Sequential',
                     ['Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',
                      'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu',
                      'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn'])
```

```python
manu = stats.loc[stats['team'] == 'Manchester United']['goals']
manc = stats.loc[stats['team'] == 'Manchester City']['goals']
liverpool = stats.loc[stats['team'] == 'Liverpool']['goals']
arsenal = stats.loc[stats['team'] == 'Arsenal']['goals']
spurs = stats.loc[stats['team'] == 'Tottenham Hotspur']['goals']
chelc = stats.loc[stats['team'] == 'Chelsea']['goals']

labels = ['Man Utd', 'Man City', 'Liverpool', 'Arsenal', 'Spurs', 'Chelsea']
colors = ['red', '#03ecfc', '#f52a3e', '#ffbace', '#f2e4ee', 'blue']

plt.pie([manu.sum(), manc.sum(), liverpool.sum(), arsenal.sum(), spurs.sum(), chelc.sum()], labels=labels, colors=colors,
        autopct='%.2f%%')
plt.title("Percentage of goals scored by the "+"Top 6")
plt.show()
```

```python
data = stats['goals']
num_bins = 5
colors = ['red', 'blue', 'green', 'orange', 'purple']

n, bins, patches = plt.hist(data, bins=num_bins)

for i in range(num_bins):
    patches[i].set_facecolor(colors[i])

plt.xlabel("Goals")
plt.ylabel("Frequency")
plt.title("Distribution of Goals Scored")
plt.show()
```
