```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/housing 2.csv')

df.head()

plt.scatter(df['total_rooms'], df['median_house_value'], color='#4de8bc', alpha=0.5)
plt.title('Total Rooms vs Median House Value')
plt.xlabel('Total Rooms')
plt.ylabel('Median House Value')
plt.show()
```

```python
Age_count = df['housing_median_age'].value_counts().sort_index()

# Bar plot
Age_count.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of People by Housing Median Age')
plt.xlabel('Housing Median Age')
plt.ylabel('Number of People')
plt.show()
```
